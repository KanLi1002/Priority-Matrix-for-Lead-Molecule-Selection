```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load data from Excel
file_path = "input.xlsx"  # Replace with your actual file path
df = pd.read_excel(file_path)

# Generate colors based on index
colors = plt.cm.viridis(np.linspace(0, 1, len(df)))

# 3D Scatter Plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

sc = ax.scatter(df['pEC50'], df['CC50'], df['T1/2'], c=colors, s=100, edgecolors='k')

# Labels
ax.set_xlabel('pEC50 (Potency)')
ax.set_ylabel('CC50 (Cytotoxicity,μM)')
ax.set_zlabel('T1/2 (Stability,min)')
ax.set_title('Lead Molecule Selection')

# Annotate points
for i, row in df.iterrows():
    ax.text(row['pEC50'], row['CC50'], row['T1/2'], row['Molecule'], fontsize=10)

# Save plot as SVG
plt.savefig("3d_scatter_plot.svg", format="svg")

plt.show()
```
